```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data
```

```python
# Compare churn count with respect to gender.
sns.countplot(x='Churn',data=data,hue='gender',palette="coolwarm_r")
```

```python
# Find out how many female senior citizens there in the dataset
sns.countplot(x='gender',data=data,hue='SeniorCitizen',palette="viridis")
```

```python
# scatter plot for Comparing 'tenure' with 'Total Charges'
sns.relplot(x="tenure", y="TotalCharges", data=data)
```

```python
# Find out which contract is preferred by the senior citizen.
```

```python
# new data with senior citizen equal 1
```

```python
data_new = data[data['SeniorCitizen'] ==1.0]
```

```python
data_new
```

```python
data_pie = data_new['Contract'].value_counts().rename_axis('Contract').reset_index(name='Contract Counts')
data_pie
```

```python
# plot a pie chart
contract_counts = [807, 190, 145]
lables = ["Month-to-month", "One year", "Two year"]
colors = ["pink", "lavender", "lightblue"]

plt.pie(contract_counts, labels=lables, autopct="%1.1f%%",colors=colors)
plt.title('The Contract preferred by Senior Citizen')
plt.show()
```
